```python
# I want a python script to open the MATH DATASET for language models to run it on OpenAI's GPT-4-Turbo API. there are json files named 1.json and so on in  ./MATH/train/algebra, and their format is:

# {
#     "problem": "Let \\[f(x) = \\left\\{\n\\begin{array}{cl} ax+3, &\\text{ if }x>2, \\\\\nx-5 &\\text{ if } -2 \\le x \\le 2, \\\\\n2x-b &\\text{ if } x <-2.\n\\end{array}\n\\right.\\]Find $a+b$ if the piecewise function is continuous (which means that its graph can be drawn without lifting your pencil from the paper).",
#     "level": "Level 5",
#     "type": "Algebra",
#     "solution": "For the piecewise function to be continuous, the cases must \"meet\" at $2$ and $-2$. For example, $ax+3$ and $x-5$ must be equal when $x=2$. This implies $a(2)+3=2-5$, which we solve to get $2a=-6 \\Rightarrow a=-3$. Similarly, $x-5$ and $2x-b$ must be equal when $x=-2$. Substituting, we get $-2-5=2(-2)-b$, which implies $b=3$. So $a+b=-3+3=\\boxed{0}$."
# }

# to do this, make a function to go through n (specified by user) files in sequential order in increasing number. Then, for each dataset, have a system prompt which instructs the API to solve the math problem in LATEX (with the final output in \\boxed{}), then compares with the answer.

```

```python
import os
from dotenv import load_dotenv

load_dotenv()

openai.api_key = os.getenv('OPENAI_API_KEY')
client = openai.Client()


def solve_math_problems(n):
    for i in range(1, n+1):
        file_path = f'./MATH/train/algebra/{i}.json'
        if not os.path.isfile(file_path):
            print(f"File {file_path} not found.")
            continue

        with open(file_path, 'r') as f:
            data = json.load(f)
            problem = data['problem']
            solution = data['solution']



            response = client.chat.completions.create(
                model="gpt-4-1106-preview",
                messages=[
                    {"role": "system", "content": "You are a helpful assistant."},
                    {"role": "user", "content": f"Solve this math problem: {problem}"},
                ],
                temperature=1,
                max_tokens=256,
                top_p=1,
                frequency_penalty=0,
                presence_penalty=0
            )
            ai_solution = response.choices[0].message.content
            print(f"Problem: {problem}")
            print(f"AI Solution: {ai_solution}")
            print(f"Actual Solution: {solution}")
            print("--------------------------------------------------")

solve_math_problems(5)  # Change the number as per your requirement
```
